```python
# *args
# allow us to pass a varaible number of non-keyword argumnet to function
def multiply(*args):
  product = 1
  for i in args:
    product = product * i
  return product
```

```python
multiply(5,6)
multiply(5,6,2)
multiply(5,6,2,4)
```

```python
# **kwargs
# **kwargs allow  us to pass any number of keyword argument
# Keyword arguments means that they contain a key-value pair, like a python dictionary

def display(**kwargs):
  for (key,value) in kwargs.items():
    print(key,'-->',value)
```

```python
display(india='delhi',srilanka='colombo',nepal='kathmandu')
```

```python
# points to remember while using *args and **kwargs
  # order of argument matter (normal->*args->**kwargs)
  # The words "args" and "kwargs" are only conventions o can use any name as per yor choice
```

```python
def square(x):
  return x**2

# HOF
def transform(f,L):
  result = []
  for i in L:
    result.append(f(i))

  print(result)


L  = [2,3,4,5,6]
transform(square,L)

transform(lambda x:x**2,L)
transform(lambda x:x**3,L)


```
