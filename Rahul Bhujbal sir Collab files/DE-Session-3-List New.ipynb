```python
# Charactrastics of Lists
   # ordered
   # Changable/Mutable
   # Hetrogrnous
   # Can have duplicate
   # dynamic
   # Item can be accessed
   # can contain objects
```

```python
# Orderd
L=[1,2,3]
L1=[3,2,1]
print(L==L1)
# Changable/Mutable
print(id(L))
L=[1,2,3,9,5]
print(id(L))
# Hetrogeneous
L = [2,5.6,10,True,'Hello',[9,8,7],2]
print(L)
```

```python
# Creating a list
# empty
L=[]
print(L)
# 1D-- Homogeneous list
L=[1,2,3,4,5,6]
print(L)
# 2D
L =[1,2,3,[4,5]]
print(L)
# 3D list
L = [[[1,2,3],[1,2]],[[5,6],[7,8]]]
print(L)
# Hetrogenous
L = [1,True,5.6,5+6j,'Hello']
print(L)
# using Conversion function
L = 'Hello'
print(L)
L = list('Hello')
print(L)
```

```python
# Accessing Items from the lists
# Indexing
L = [1,2,3,4,5,6]
print(L[0])
print(L[3])
#print(L[30])
print(L[-1])
print(L[-3])

L = [1,2,3,[4,5]]
print(L[1])
print(L[3])
print(L[3][0])

L = [[[1,2,3],[1,2]],[[5,6],[7,8]]]
print(L[1])
print(L[1][0])
print(L[1][0][1])

# Slicing
L = [1,2,3,4,5,6]
print(L[0:3])
print(L[-3:])
print(L[0::2])
print(L[-5:-2:2])
print(L[::-1])# Reverse the list
```

```python
# Adding items into the list
#append
L = [1,2,3,4,5]
print(L)
print(id(L))
L.append(10)
print(L)
print(id(L))
L.append(True)
print(L)
print(id(L))
L = [1,2,3,4,5]
print(L)
L.append([10,20,30])
print(L)
print(L)
L.extend([10,20,30])
print(L)

L = [1,2,3,4,5]
print(L)
L.append(10)
print(L)
L.append([10,20,30])
print(L)
L.extend([10,20,30])
print(L)
```

```python
L = [1,2,3,4,5]
print(L)
L.insert(1,100)
print(L)
```

```python
# Editing the items
L = [1,2,3,4,5,6]
print(L)
L[-1] = 600
print(L)
L[1:4]=[200,300,400]
print(L)
```

```python
# Deleting items from a list
# del
L = [1,2,3,4,5]
#del L
print(L)

del L[-1]
print(L)
# remove
L = [1,2,3,4,5]
L.append(4)
print(L)
L.remove(4)
print(L)

# pop
L = [1,2,3,4,5]
L.pop(0)
print(L)
# clear
L = [1,2,3,4,5]
L.clear()
print(L)
```

```python
# Opearation on the list

# Arithmetic (+,*)

L1 = [1,2,3,4]
L2 = [5,6,7,8]

print(L1+L2)
print(L1*3)

# Membership operator
L1 = [1,2,3,4]
L2 = [5,6,7,8]
print(5 in L2)
print(5 not in L2)

L1 = [1,2,3,4,[5,6]]
L2 = [5,6,7,8]
print([5,6] in L1)
print([5,6] not in L1)

# Loop
L1 = [1,2,3,4,[5,6]]
L2 = [5,6,7,8]


for i in L1:
  print(i)

for i in L2:
  print(i)

L3 = [[[1,2,3],[1,2]],[[5,6],[7,8]]]

for i in L3:
  print(i)
```

```python
# count

L = [1,2,3,1,3,4,1,5]
print(L.count(5))
print(L.count(1))
print(L.count(3))
```

```python
L = [1,2,5,7,0]
print(id(L))
L.reverse() # permanently reverse the list
print(id(L))
print(L)

S = 'Rahul'
print(S[::-1])
print(id(S))
s1 = reversed(S)
print(id(s1))
for i in s1:
  print(i)
```

```python
# sort VS sorted
L = [2,1,5,7,0]
print(L)
print(sorted(L))
print(L)
L.sort()
print(L)

```

```python
# COPY -> shallow copy
L = [2,1,5,7,0]
print(L)
print(id(L))
L1=L.copy()
print(L1)
print(id(L1))
```

```python
# List Comprehension
# List Comprehension provide a easy way/choice of creating a list

# newlist = [expression for item in itrable if condition == True]

# Add 1 to 10 numbers in list
L = []
for i in range(1,11):
  L.append(i)
print(L)

L = [i for i in range(1,11)]
print(L)

# Scalar multiplicaton on vector

v = [2,3,4]
s = -3
print(v*-s)

#output [-6,-9,-12]

x = []
for i in v:
  x.append(i*s)
print(x)

x = [s*i for i in v]
print(x)


L = [1,2,3,4,5]
#output [1,4,3,16,25]

s = []
for i in L:
  s.append(i**2)
print(s)

s = [i**2 for i in L]
print(L)

# print all the numbers divisible by 5 in the range of 1 to 50

L = []
for i in range(1,51):
  if i%5 == 0:
    L.append(i)
print(L)

L = [i for i in range(1,51) if i%5==0]
print(L)
```

```python
# find the language which start with letter p

language = ['java','python','php','c','plsql']
plang = []
for i in language:
  if i.startswith('p'):
    plang.append(i)
print(plang)

plang = [i for i in language if i.startswith('p')]
print(plang)

# Advantage

# More time-efficient and space-efficient
# Require fewer line of code
# Tranform itrative statement into formula
```

```python
# Ways to traverse a list
# itemwise
# indexwise

L = [1,2,3,4]
for i in L:
  print(i)

L = [1,2,3,4]
for i in range(0,len(L)):
  print(i)

L = [1,2,3,4]
for i in range(0,len(L)):
  print(L[i])
```
