```python
# Else
try:
  f = open('sample.txt','r')
except FileNotFoundError as e:
  print(e.args)
except Exception as e:
  print(e.args)
else:
  print(f.read())
```

```python
# Finally
try:
  f = open('sample.txt','r')
  print(f.read())
except FileNotFoundError as e:
  print(e.args)
except Exception as e:
  print(e.args)
finally:
  print('Always finally block will execute')
```

```python
class Bank:
  def __init__(self,balance):
    self.balance = balance
  def withdraw(self,amount):
    if amount < 0:
      raise Exception('Amount cannot be negative')
    if self.balance < amount:
      raise Exception('Insufficient Balance')

    self.balance = self.balance - amount
```

```python
b = Bank(10000)
try:
  b.withdraw(20000)
except Exception as e:
  print(e.args)
else:
  print(b.balance)
```
