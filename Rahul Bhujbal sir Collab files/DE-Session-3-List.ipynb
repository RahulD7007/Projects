```python
# What are the Lists
   # List is a data type where we can store multiple item under one name. more technically,
   # we can say lists acts like dynamic array which means we can add more items on the fly.

# Few differences b/w Array Vs Lists
  # Fixed vs Dynamic size
  # Convenience -->  Heterogeneous
  # Speed of execution
  # Memory
```

```python
# How Lists are stored in memory

a = 2
print(id(2))
print(id(a))

L = [1,2,3]
print(id(L))
print(id(L[0]))
print(id(L[1]))
print(id(L[2]))
print(id(1))
print(id(2))
```

```python
# Characteristics of Lists
  # Ordered
  # Changeable/Mutable
  # Heterogeneous
  # Can have duplicate
  # Dynamic
  # Item can be accessed
  # can contain objects
```

```python
# Ordered
L =  [1,2,3]
L1 = [3,2,1]
print(L==L1)

# Changeable/Mutable
print(id(L))
L = [1,2,3,9,5]
print(id(L))

# Heterogeneous
L = [2,5.6,10,True,'Hello',[9,8,7],2]
print(L)
```

```python
# Creating a list
# Empty
L = []
print(L)

# 1D -> Homogeneous List
L = [1,2,3,4,5,6]
print(L)

# 2D ->
L = [1,2,3,[4,5,6]]
print(L)

# 3D ->
L = [[[1,2,3],[1,2]],[[5,6],[7,8]]]
print(L)

# Heterogeneous
L = [1,True,5.6,5+6j,'Hello']
print(L)

# By using conversion function
L = 'Hello'
print(L)
L = list('Hello')
print(L)

```

```python
# Accessing items from the lists
# Indexing
L = [1,2,3,4,5,6]
print(L[0])
print(L[3])
# print(L[30])
print(L[-3])
print(L[-1])

L = [1,2,3,[4,5]]
print(L[1])
print(L[3])
print(L[3][0])

L = [[[1,2,3],[1,2]],[[5,6],[7,8]]]
print(L[1])
print(L[1][0][0])

# Slicing
L = [1,2,3,4,5,6]
print(L[0:3])
print(L[-3:])
print(L[0::2])
print(L[-5:-2:2])
print(L[::-1]) # Reverse the list
```

```python
# Adding items into the list

# Append
L = [1,2,3,4,5]
print(L)
print(id(L))

L.append(10)
print(L)
print(id(L))

L.append(True)
print(L)
print(id(L))

L = [1,2,3,4,5]
print(L)
L.append([10,20,30])
print(L)
print(id(L))

L.extend([10,20,30])
print(L)
print(id(L))

L = [1,2,3,4,5]
print(L)
L.append(10)
print(L)
L.append([10,20,30])
print(L)
print(id(L))
L.extend([10,20,30])
print(L)
print(id(L))
```

```python
L = [1,2,3,4,5]
print(L)
L.append('pune')
print(L)

L.extend('pune')
print(L)
```

```python
L = [1,2,3,4,5]
print(L)
L.insert(1,100)
print(L)
```

```python
# Editing the items
L = [1,2,3,4,5]
print(L)
L[-1] = 600
print(L)
L[1:4] = [200,300,400]
print(L)
```

```python
# Deleting the items from a lists
L = [1,2,3,4,5]
# del L
print(L)

del L[-1]
print(L)

# remove
L = [1,2,3,4,5]
L.remove(3)
print(L)

```

```python
# operation on the lists

# remove
L = [1,2,3,4,5,4]
L.remove(4)
print(L)

# pop
L = [1,2,3,4,5]
L.pop(0)
print(L)

# clear
L = [1,2,3,4,5]
L.clear()
print(L)
```

```python
L = [1, 4, 3, 4, 5, 4, 2, 4]
while 4 in L: L.remove(4)
print(L)

L = list(dict.fromkeys(L))
print(L)
```
