```python
class EmployeesManager:
  def __init__(self):
    self.employees = []

  def menu(self):
    user_input = input(""" Ente your choice
                       1. press 1 to add employee
                       2. press 2 to desplay employee
                       3. press 3 to remove all employee
                       4. Anything else to exist""")
    if user_input == '1':
      # add employee
      self.add_employee()

    elif user_input == '2':
      # display employee
      self.display_employee()

    elif user_input == '3':
      # remove all employee
      self.remove_all_employee()

    else:
      exit()

  def add_employee(self):
    print('Enter employee details')
    name = input('Enter employee name')
    self.employees.append(name)
    age = input('Enter employee age')
    self.employees.append(age)
    salary = input('Enter employee salary')
    self.employees.append(salary)
    print('Employee added successfully')
    self.menu()

  def display_employee(self):
    if len(self.employees) == 0:
      print('No employee found')
      self.menu()
    else:
      for emp in self.employees:
        print(emp)
    self.menu()

  def remove_all_employee(self):
    for i in self.employees:
      print('Deleting employees')
      self.employees.clear()
      print('All employees deleted successfully')
    self.menu()
```
