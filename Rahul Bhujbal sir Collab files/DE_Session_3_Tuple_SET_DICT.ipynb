```python
# Tuple
  # A tuple in python is a similar to list. The difference between the list and tuple is that we can not change the item of tuple
  # Once it is assigned whereas we can change the element of list.
  # In short, tuple is immutable list

# Charactrestics
  # orderd
  # Unchangabel
  # duplicates are allowed

# Plan of attack
  # creating a tuple
  # Accessiong items
  # Editing items
  # Adding items
  # deleting items
  # operation on tuple
  # Tuple functions
```

```python
# Creating Tuple

# empty
t1 = ()
print(t1)

# Create a tuple with single item
L = [2]
print(L)
print(type(L))
t2 = (2)
print(t2)
print(type(t2))

t3 = ('Rahul')
print(t3)
print(type(t3))

t4 = (2,)
print(t4)
print(type(t4))

# Homogenious tuple
t5 = (1,2,3,4,5)
print(t5)

# hetrogenious tuple
t6 = (5,8.56,'hello',True,[1,2,3,4])
print(t6)

t7 = (1,2,3,(4,5))
print(t7)

# type conversion

t8 = tuple('Rahul')
print(t8)
```

```python
# Accessing items
# Indexing

t = (1,2,3)
print(t)
print(t[0])
print(t[-1])
#print(t[-6])

# Slicing

t2 = (1,2,3,4)
print(t2[0:4])
print(t2[0:4:2])
print(t2[-3:-1])
print(t2[::-1])

t3 = (1,2,3,(4,5))
print(t3)
print(t3[-1][0])
```

```python
# Editing-> Not able to edit tuple becouse its an immutable
L = [1,2,3,4]
print(L)
L[0]=100
print(L)
t3 = (1,2,3,4)
print(t3)
t3[0]=100
print(t3)
```

```python
# Adding items
t3 = (1,2,3,4)
print(t3)
t3.append(5)
print(t3)
```

```python
# Deleting

t3 = (1,2,3,4)
print(t3)
del t3
#print(t3)
```

```python
# Tuple Functions

# len/sum/min/max/sorted

t = (8,7,23,6,45)
print(len(t))
print(sum(t))
print(min(t))
print(max(t))
print(sorted(t))
print(sorted(t,reverse=True))

# count

t = (1,2,1,6,4,5,4,8,5,1)

print(t.count(1))
print(t.count(5))

# Index

print(t.index(5))
#print(t.index(12))
```

```python
# Diff between Lists and Tuple
# Syntax
# Mutablity
# Speed
# Memory
# Bult in functionality
# Error Prone
# Usability
```

```python
# Speed
import time
L = list(range(100000000))
T = tuple(range(100000000))

start = time.time()
for i in L:
  i*5
print('List Time: ',time.time()-start)

start = time.time()
for i in T:
  i*5
print('Tuple Time: ',time.time()-start)
```

```python
# Memory
import sys
L = list(range(100000000))
T = tuple(range(100000000))
print('List Size :', sys.getsizeof(L))
print('Tuple Size :', sys.getsizeof(T))
```

```python
# Sets
# A set is an unorderd collection of items. Every set element is unique (no duplicate) and must be immutable.
# Set can also be used to perform mathematical operations like union, intesection.

# Characterstics
# Unorderd
# Mutable/immutable
# No duplicate
# Cant contain mutable datatype
```

```python
# Creating Set

# empty set
s = {}
print(s)
print(type(s))

s = set()
print(s)
print(type(s))

s1 = {1,2,3}
print(s1)

#s3 = { 1,2,3,[4,5]} # Can not have mutable datatype
#print(s3)

s4 = { 1,2,3,{4,5}}# Can not have mutable datatype
print(s4)
```

```python
# homo and hetro
s = {1,2,3,4,5}
print(s)

s2 = {1,2,3,4.5,'hello'}
print(s2)

s3 = {(1,2,3),4,5}
print(s3)

# Using type conversion

s4 = set([1,2,3,4,5])
print(s4)

# Duplicate not allowd
s5 = {1,2,3,4,5,1,2,3}
print(s5)

# Unorderd
s1 = {1,2,3}
s2 = {3,2,1}
print(s1==s2)
```

```python
# Accessing items does not work becouse its unorderd

s1 = {1,2,3,4}
print(s1[0])
```

```python
# Set operation
s1 = {1,2,3,4}
s2 = {3,4,5,6}

# Union(|)
print(s1|s2)
print(s1.union(s2))

# Intersection(&)
print(s1&s2)
print(s1.intersection(s2))

# difference(-)

print(s1-s2)
print(s2-s1)
print(s1.difference(s2))
print(s2.difference(s1))

# Symmetric difference(^)
print(s1^s2)
print(s1.symmetric_difference(s2))
```

```python
# Membership
print(1 in s2)
print(1 not in s2)
```

```python
# isdisjoint/issubset/issuperset

#isdisjoint
s1 = {6,7}
s2 = {3,4}
print(s1.isdisjoint(s2)) # if all items are different in set then it will return true other wise it will false

#issubset

s3= {1,2,3,4}
s4 = {1,2}

print(s3.issubset(s4))
print(s4.issubset(s3))

# issuperset
print(s3.issuperset(s4))
print(s4.issuperset(s3))
```

```python
# Frozenset
# Frozenset is just an immutable version of python set object

# create
fs = frozenset([1,2,3,4])
print(fs)
print(type(fs))

fs1 = frozenset((1,2,3,4))
print(fs1)
print(type(fs1))

fs2 = frozenset({1,2,3})
print(fs2)
print(type(fs2))
```

```python
# Accessing items
my_dict = {'name':'Rahul','age':36,'gender':'Male'}
print(my_dict['name'])
print(my_dict['age'])
print(my_dict['gender'])

s = {
    'name':'Rahul',
    'college': 'bit',
    'sem' : 4,
    'Subjects': {
        'math': 89,
        'english': 75,
        'python': 95
    }
}

print(s)
print(s['Subjects']['math'])

# get
print(my_dict.get('name'))
print(my_dict.get('age'))
print(my_dict.get('gender'))
```

```python
# Adding key-value pair
my_dict = {'name':'Rahul','age':36}
print(my_dict)
my_dict['gender'] = 'Male'
print(my_dict)
my_dict['weight'] = 61
print(my_dict)

s = {
    'name':'Rahul',
    'college': 'bit',
    'sem' : 4,
    'Subjects': {
        'math': 89,
        'english': 75,
        'python': 95
    }
}

print(s)
s['Subjects']['oracle'] = 79
print(s)
```

```python
# Remove key-value pair
my_dict = {'name':'Rahul','age':36,'gender':'Male','weight':61}
print(my_dict)
# pop
my_dict.pop('gender')
print(my_dict)

# popitem
my_dict = {'name':'Rahul','age':36,'gender':'Male','weight':61}
print(my_dict)
my_dict.popitem()
print(my_dict)
my_dict.popitem()
print(my_dict)

# del
my_dict = {'name':'Rahul','age':36,'gender':'Male','weight':61}
print(my_dict)
del my_dict['gender']
print(my_dict)
s = {
    'name':'Rahul',
    'college': 'bit',
    'sem' : 4,
    'Subjects': {
        'math': 89,
        'english': 75,
        'python': 95
    }
}

print(s)
del s['Subjects']['math']
print(s)

# clear
my_dict = {'name':'Rahul','age':36,'gender':'Male','weight':6}
print(my_dict)
my_dict.clear()
print(my_dict)
```

```python
# Editing key-value pair

s = {
    'name':'Rahul',
    'college': 'bit',
    'sem' : 4,
    'Subjects': {
        'math': 89,
        'english': 75,
        'python': 95
    }
}

print(s)
s['sem']=5
print(s)
s['Subjects']['english']=55
print(s)
```

```python
# Dictionary operations
# membership
s = {
    'name':'Rahul',
    'college': 'bit',
    'sem' : 4,
    'Subjects': {
        'math': 89,
        'english': 75,
        'python': 95
    }
}

print(s)
print('Rahul' in s) # it will get false becouse in dictionary membership work on keys only not on values
print('name' in s)

# Iteration

d = {'name':'Rahul','age':36,'gender':'Male','weight':61}
for i in d:
  print(i)

for i in d:
  print(i,d[i])
```

```python
# Dictionary function
# len/sorted
d = {'name':'Rahul','age':36,'gender':'Male','weight':61}
print(len(d))
print(sorted(d))
print(sorted(d,reverse=True))
# min/max
print(min(d))
print(max(d))

# items/keys/values
d = {'name':'Rahul','age':36,'gender':'Male','weight':61}
print(d)

print(d.items())
print(d.keys())
print(d.values())

# update
d1 = {1:2,3:4}
d2 = {5:6,7:8}
d1.update(d2)
print(d1)
print(d2)
```
