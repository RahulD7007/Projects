```python
# Menu driven calculator
fnum = int(input('Enter first number: '))
snum = int(input('Enter second number: '))

operation = input('Enter the operation: ')
if operation == '+':
  print(fnum+snum)
elif operation == '-':
  print(fnum-snum)
elif operation == '*':
  print(fnum*snum)
else:
  print(fnum/snum)
```

```python
# while loop- program to print table

num = int(input('Enter the number: '))

i = 1
while i < 11:
  print(num*i)
  i = i + 1
```

```python
# while loop with else

x = 1
while x < 3:
  print(x)
  x = x + 1
else:
  print('Limit cross')
```

```python
import random
jackpot = random.randint(1,100)
guess = int(input('Enter jackpot number: '))
counter = 1
while guess != jackpot:
  if guess < jackpot:
    print('Wrong! Please try with higher number')
  else:
    print('Wrong! Please try with lower number')
    guess = int(input('Enter jackpot number: '))
    counter = counter + 1
else:
  print('Congratulations!!! You are winner at', counter)
```

```python
# nested for loop

for i in range(1,5):
  for j in range(1,5):
    print(i,j)

```

```python
# for pattern for loop

rows = int(input('Enter number of rows: '))
for i in range(1,rows+1):
  for j in range(1,i+1):
    print('*',end = '')
  print()

```

```python
# for inverted pattern for loop

rows = int(input('Enter number of rows: '))
for i in range(5,0,-1):
  for j in range(1,i+1):
    print('*',end = '')
  print()
```

```python
# Loop control statement
# Break
# Continue
# Pass
```

```python
# Break

for i in range(1,10):
  if i == 5:
    break
  print(i)
```

```python
# Continue  (ye bahut useful nahi hai)

for i in range(1,10):
  if i == 5:
    continue
  print(i)
```

```python
# pass

for i in range(1,10):
  pass
```
