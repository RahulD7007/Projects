```python
# Method overriding
class phone:
  def __init__(self,price,brand,camera):
    self.price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
   print("Buying a phone")

class smartPhone(phone):
  def buy(self):
    print("Buying a smart phone")

s = smartPhone(20000,"Apple",20)
s.buy()
```

```python
# Super Keyword
class phone:
  def __init__(self,price,brand,camera):
    self.price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
   print("Buying a phone")

class smartPhone(phone):
  def buy(self):
    print("Buying a smart phone")
    # Synatx to call parent method
    super().buy()

s = smartPhone(20000,"Apple",20)
s.buy()
```

```python
# super -> constructor
class phone:
  def __init__(self,price,brand,camera):
    self.__price = price
    self.brand = brand
    self.camera = camera
    print("Inside Phone Constructor")


class smartPhone(phone):
  def __init__(self,price,brand,camera,os,ram):
    print("Inside Smart Phone Constructor")
    super().__init__(price,brand,camera)
    self.os = os
    self.ram = ram
s = smartPhone(200000,"Samsung",15,"Andriod",16)
print(s.os)
print(s.brand)
print(s.price)
```

```python
# Single
class phone:
  def __init__(self,price,brand,camera):
    self.price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
   print("Buying a phone")

class smartPhone(phone):
  pass

s = smartPhone(20000,"Apple",20).buy()
```

```python
# Multilevel
class product:
  def review(self):
    print("Product customer review")
class phone(product):
  def __init__(self,price,brand,camera):
    print("Inside phone constructor")
    self.price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
   print("Buying a phone")

class smartPhone(phone):
  pass


s = smartPhone(20000,"Apple",20)
s.buy()
s.review()
```

```python
# Hirerachical
class product:
  def review(self):
    print("Product customer review")
class phone(product):
  def __init__(self,price,brand,camera):
    print("Inside phone constructor")
    self.price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
   print("Buying a phone")

class smartPhone(phone):
  pass

class futurephone(phone):
  pass


s = smartPhone(20000,"Apple",20)
f = futurephone(70000,"Samsung",12)
s.buy()
f.buy()
```

```python
# Method overloading is a concept where we can write more than 1 menthod inside a single class
# Method overloading does not work in python

class shape:
  def area(self,r):
    return 3.14*r*r
  def area(self,l,b):
    return l*b

s = shape()
print("Area of circle",s.area(4))
```

```python
# We can implement method overloading using following way
class shape:
  def area(self,a,b=0):
    if b==0:
      return 3.14*a*a
    else:
      return a*b

s = shape()
print("Area of circle",s.area(4))
print("Area of Rectangle",s.area(2,3))
```
