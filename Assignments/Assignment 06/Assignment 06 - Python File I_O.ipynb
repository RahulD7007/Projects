### 2. Write a Python program to read first n lines of a file.

```python
def file_read_from_head(fname, nlines):
  from itertools import islice
  with open(fname) as f:
    for line in islice(f, nlines):
      print(line)

file_read_from_head(r"/content/Text.txt", 2)
```

### 4. Write a Python program to read last n lines of a file.

```python
import sys
import os

def file_read_from_tail(fname,lines):
  bufsize = 8192
  fsize = os.stat(fname).st_size
  iter = 0
  with open(fname) as f:
    if bufsize > fsize:
      bufsize = fsize-1
      data = []
      while True:
        iter +=1
        f.seek(fsize-bufsize*iter)
        data.extend(f.readlines())
        if len(data) >= lines or f.tell() == 0:
          print(''.join(data[-lines:]))
          break

file_read_from_tail('Text.txt',2)
```

### 7. Write a python program to find the longest words.

```python
def longest_word(filename):
  with open(filename, 'r') as infile:
    words = infile.read().split()

  max_len = len(max(words, key=len))
  return [word for word in words if len(word) == max_len]

print(longest_word('Text.txt'))
```

### 9. Write a Python program to count the frequency of words in a file.

```python
from collections import Counter

def word_count(fname):
  with open(fname) as f:
    return Counter(f.read().split())

print("Number of words in the file :",word_count("Text.txt"))
```

### 10. Write a Python program to get the file size of a plain file.

```python
def file_size(fname):
  import os
  statinfo = os.stat(fname)
  return statinfo.st_size

print("File size in bytes of a plain file: ",file_size("Text.txt"))
```

### 13. Write a Python program to read a random line from a file.

```python
import random
def random_line(fname):
  lines = open(fname).read().splitlines()
  return random.choice(lines)

print(random_line('Text.txt'))
```

### 15.Write a Python program that takes a text file as input and returns the number of words of a given text file.

```python
# Note: Some words can be separated by a comma with no space.
def count_words(filepath):
  with open(filepath) as f:
    data = f.read()
    data.replace(",", " ")
    return len(data.split(" "))

print(count_words("/content/Text.txt"))
```

### 16. Write a Python program to read a file line by line store it into an array.

```python
def file_read(fname):
  content_array = []
  with open(fname) as f:
    # Content_list is the list that contains the read lines.
    for line in f:
      content_array.append(line)
    print(content_array)

file_read('Text.txt')
```

### 18. Write a Python program to remove newline characters from a file.

```python
def remove_newlines(fname):
  flist = open(fname).readlines()
  return [s.rstrip('\n') for s in flist]

print(remove_newlines("Text.txt"))
```
